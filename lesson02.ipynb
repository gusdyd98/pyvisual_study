# 초기화 코드

```python
%matplotlib inline

from matplotlib import pyplot as plt
import pandas as pd

plt.rc('axes', unicode_minus=False)
plt.rc('font', family='Malgun Gothic')

plt.style.use('fast')
```

```python
x = [0, 1, 2, 3]
y = [1, 4, 9, 16]

df=pd.DataFrame(y, index=x)
ax=df.plot(title='Line Plot', figsize=(10,5), grid=True)
ax.set_xlabel('X축')
ax.set_ylabel('Y축')
```

## 가용한 폰트 목록

```python
import os
from matplotlib.font_manager import fontManager
import pandas as pd
```

```python
font_list=[]
for font in fontManager.ttflist:
    font_list.append({
        'name': font.name,
        'fname': os.path.basename(font.fname),
        'style': font.style,
    })
    
df = pd.DataFrame(font_list).set_index('name')
print(df.shape)
df.head()
```
